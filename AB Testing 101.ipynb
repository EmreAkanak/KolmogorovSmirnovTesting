## Test(ing) Examples: 

### Dependent Pairs A/B Test(ing)Â 

```python

# Classical Dependent Pairs A/B Test(ing)

from scipy import stats as st
import numpy as np

before = [157, 114, 152, 355, 155, 513, 299, 268, 164, 320, 
                    192, 262, 506, 240, 364, 179, 246, 427, 187, 431, 
                    320, 193, 313, 347, 312, 92, 177, 225, 242, 312]

after = [282, 220, 162, 226, 296, 479, 248, 322, 298, 418, 
                 552, 246, 251, 404, 368, 484, 358, 264, 359, 410, 
                 382, 350, 406, 416, 438, 364, 283, 314, 420, 218]

alpha = 0.05  # critical statistical significance level


print('H0: The effect of the change is not significant, the after results are the same with/as before results')
print()
print('HA: The effect of the change is significant, the after results are not the same with/as before results')


results = st.ttest_rel(before, after)

print('p-value: ', results.pvalue)





if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis") 

    
print()

print('Note: Result is based on {:.2%} treshold, which means calculation is based on {:.2%} Confidence Interval'.format(alpha, (1-alpha)))    
```

### Independent Pairs A/B Test(ing)

```python
# Classical Independent Pairs A/B Test(ing)

from scipy import stats as st
import numpy as np

control = [157, 114, 152, 355, 155, 513, 299, 268, 164, 320, 
                    192, 262, 506, 240, 364, 179, 246, 427, 187, 431, 
                    320, 193, 313, 347, 312, 92, 177, 225, 242, 312]

treatment = [282, 220, 162, 226, 296, 479, 248, 322, 298, 418, 
                 552, 246, 251, 404, 368, 484, 358, 264, 359, 410, 
                 382, 350, 406, 416, 438, 364, 283, 314, 420, 218]

alpha = 0.05  # critical statistical significance level

print('H0: The Treatment Group results are the same with the Control Group Results')
print()
print('HA: The Treatment Group results are not the same with the Control Group Results')

print()

results = st.ttest_ind(control, treatment)


print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis") 

print()

print('Note: Result is based on {:.2%} treshold, which means calculation is based on {:.2%} Confidence Interval'.format(alpha, (1-alpha)))    
```
